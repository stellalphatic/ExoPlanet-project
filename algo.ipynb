```python
import csv

# Define Earth's benchmark figures for habitability
earth_benchmark = {
    'distance': 1.0,  # astronomical units (AU)
    'density': 5.5,  # g/cm^3
    'temp': 288,  # Kelvin
    'mass': 1,  # Earth masses (M)
    'radius': 1,  # Earth radii (R)
    'flux': 1.0,  # flux of the star
    'period': 365,  # orbital period in Earth days
    'eccentricity': 0.0,  # eccentricity of the orbit
    'gravity': 1.0,  # surface gravity in g
    'esi_habitable': True,  # habitability according to ESI
    'atmospheric_pressure': 1013,  # mbar
    'atmospheric_composition': 'N2/O2',  # atmospheric composition
    'surface_water': True,  # presence of surface water
    'magnetic_field': True,  # presence of magnetic field
    'geological_activity': True,  # presence of geological activity
}

# Define maximum allowed differences for survival
max_diff = {
    'distance': 0.5,  # AU
    'density': 1.0,  # g/cm^3
    'temp': 50,  # Kelvin
    'mass': 0.5,  # M
    'radius': 0.5,  # R
    'flux': 0.5,  # flux of the star
    'period': 100,  # Earth days
    'eccentricity': 0.5,  # eccentricity of the orbit
    'gravity': 0.5,  # g
    'atmospheric_pressure': 500,  # mbar
    'atmospheric_composition': 0.5,  # similarity to Earth's atmospheric composition
    'surface_water': True,  # presence of surface water
    'magnetic_field': True,  # presence of magnetic field
    'geological_activity': True,  # presence of geological activity
}

# Load exoplanet data from CSV file
exoplanets = []
with open('ExoPlannetData.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        exoplanets.append(row)

for exoplanet in exoplanets:
    # Calculate differences between exoplanet and Earth's benchmark figures
    diff = {
        'distance': abs(float(exoplanet.get('distance', 0)) - earth_benchmark['distance']),
        'density': abs(float(exoplanet.get('density', 0)) - earth_benchmark['density']),
        'temp': abs(float(exoplanet.get('temp', 0)) - earth_benchmark['temp']),
        'mass': abs(float(exoplanet.get('mass', 0)) - earth_benchmark['mass']),
        'radius': abs(float(exoplanet.get('radius', 0)) - earth_benchmark['radius']),
        'flux': abs(float(exoplanet.get('flux', 0)) - earth_benchmark['flux']),
        'period': abs(float(exoplanet.get('period', 0)) - earth_benchmark['period']),
        'eccentricity': abs(float(exoplanet.get('eccentricity', 0)) - earth_benchmark['eccentricity']),
        'gravity': abs(float(exoplanet.get('gravity', 0)) - earth_benchmark['gravity']),
        'atmospheric_pressure': abs(float(exoplanet.get('atmospheric_pressure', 0)) - earth_benchmark['atmospheric_pressure']),
        'atmospheric_composition': 1 - (exoplanet.get('atmospheric_composition', '') == earth_benchmark['atmospheric_composition']),
        'surface_water': not (exoplanet.get('surface_water', False) == earth_benchmark['surface_water']),
        'magnetic_field': not (exoplanet.get('magnetic_field', False) == earth_benchmark['magnetic_field']),
        'geological_activity': not (exoplanet.get('geological_activity', False) == earth_benchmark['geological_activity']),
    }

    # Check if exoplanet is habitable based on maximum allowed differences
    habitable = True
    for key, value in diff.items():
        if key in ['atmospheric_composition', 'surface_water', 'magnetic_field', 'geological_activity']:
            if value != 0:
                habitable = False
                break
        elif value > max_diff[key]:
            habitable = False
            break

    # Calculate habitability score
    habitability_score = {key: max(0, max_diff[key] - value) for key, value in diff.items() if key in max_diff}

    # Calculate final habitability score as a percentage
    final_habitability_score = sum(habitability_score.values()) / sum(max_diff.values()) * 100

    # Print results
    print(f"Exoplanet: {exoplanet.get('Name')}")
    print(f"Habitable: {habitable}")
    print(f"Differences:")
    for key, value in diff.items():
        print(f"  {key}: {value:.2f}")
    print(f"Habitability Score: {final_habitability_score:.2f}%")
    print()
```
